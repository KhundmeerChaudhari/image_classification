```python
dataset = []
for image in glob.glob("/content/dataset_full/*"+'/*.jpg'):
    defect_name = image.split('/')[-2]
    image = cv2.imread(image)
    rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = cv2.blur(rgb_image, (3, 3))


    image = cv2.resize(image, (200, 200))
    #plt.imshow(image)

    im = list(np.ravel(image))
    row = []
    for i in im:
        row.append(int(i))
    defect = defect_name
    row.append(defect)
    dataset.append(row)

df = pd.DataFrame(dataset)
df.fillna(0, inplace=True)
df.to_csv("/content/classify2.csv", index=None)
```

```python
# Kernel mlp

# Importing the libraries
# Importing the dataset
dataset = pd.read_csv("/content/classify2.csv")
print(dataset.head())
print(dataset.shape)
X = dataset.iloc[:, :30000].values
y = dataset.iloc[:, -1].values

import sklearn


# X = df.iloc[:7500]
# y = df.iloc[:-1]

# print(X.head())
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 0)

# Feature Scaling
# from sklearn.preprocessing import StandardScaler
# sc = StandardScaler()
# X_train = sc.fit_transform(X_train)
# X_test = sc.transform(X_test)

# Fitting Kernel SVM to the Training set
from sklearn.svm import SVC
classifier = SVC(kernel = 'rbf', random_state = 0,C=0.5,degree=1)
classifier.fit(X_train, y_train)
loss=classifier.score(X_test,y_test)
print(loss)

# Predicting the Test set results
y_pred = classifier.predict(X_test)

# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
# Kernel mlp

# Importing the libraries
# Importing the dataset
dataset = pd.read_csv("/content/classify.csv")
print(dataset.head())
print(dataset.shape)
```

```python
X = dataset.iloc[:, :30000].values
y = dataset.iloc[:, -1].values
```

```python
import sklearn


# X = df.iloc[:7500]
# y = df.iloc[:-1]

# print(X.head())
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 0)

# Feature Scaling
# from sklearn.preprocessing import StandardScaler
# sc = StandardScaler()
# X_train = sc.fit_transform(X_train)
# X_test = sc.transform(X_test)

# Fitting Kernel SVM to the Training set
from sklearn.svm import SVC
classifier = SVC(kernel = 'rbf', random_state = 0,C=0.5,degree=1)
classifier.fit(X_train, y_train)
loss=classifier.score(X_test,y_test)
print(loss)

# Predicting the Test set results
y_pred = classifier.predict(X_test)

# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```
